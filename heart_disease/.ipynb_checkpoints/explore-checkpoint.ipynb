## PCA

```python
y = df.target.values
df_X = df.drop(['target'], axis=1)
X = df_X.values
feature_names = list(df_X.columns)
```

```python
from sklearn.preprocessing import Normalizer

transformer = Normalizer().fit(X)
transformer.transform(X)
X_norm = transformer.transform(X)
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.2, random_state=42)
# print(f'Frauds in test set: {y_test.sum()}\nFrauds in train set: {y_train.sum()}')
X_train = X_norm
y_train = y
```

```python
X_train.shape
```

```python
from sklearn.decomposition import PCA

n_components = 5
pca = PCA(n_components=n_components)
pca_transform = pca.fit(X_train)

X_pca = pca_transform.transform(X_train)

X_pos = X_pca[y_train==0]
X_anomaly = X_pca[y_train==1]
```

```python
for i, j in itertools.combinations(range(n_components), 2):

      
    fig, ax = plt.subplots(1, figsize=(6,6))
    ax.scatter(X_pos[:, i], X_pos[:, j], alpha=1, marker='H', label='operazioni standard')
    ax.scatter(X_anomaly[:, i], X_anomaly[:, j], alpha=1, color='red', label='anomalie')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend()
    ax.set_title(f'i={i}, j={j}')
    plt.axis('off')
    plt.show()
# fig.savefig('../plots/anomalie.png', bbox_inches='tight', dpi=300)
```
