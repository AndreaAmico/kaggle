```python
from keras import models
# from keras.applications import Xception
# from keras.applications import VGG16
from keras.applications import VGG19


from keras import layers
    

conv_base = VGG19(weights = 'imagenet',
                    include_top = False,
                    input_shape = (224, 224, 3))


conv_base.trainable=False

model = models.Sequential()

for layer in conv_base.layers[:-5]:
    layer.trainable = False
    model.add(layer)
    
    
model.add(layers.Flatten())
#model.add(layers.Dense(20, activation='relu'))
#model.add(layers.Dropout(rate=0.1))
# model.add(layers.GlobalAveragePooling2D())
model.add(layers.Dense(200, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))


```

```python
from keras import models
from keras import layers


img_input = layers.Input(shape=(224, 224, 3))
block1_conv1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same', name='block1_conv1')(img_input)
block1_pool = layers.MaxPooling2D((2, 2), strides=(2, 2), name='block1_pool')(block1_conv1)

block2_conv1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='block2_conv1')(block1_pool)
block2_conv2 = layers.SeparableConv2D(64, (3, 3), activation='relu', padding='same', name='block2_conv2')(block2_conv1)
block2_conv3 = layers.SeparableConv2D(64, (3, 3), activation='relu', padding='same', name='block2_conv3')(block2_conv2)
block2_pool = layers.MaxPooling2D((2, 2), strides=(2, 2), name='block2_pool')(block2_conv3)

block3_conv1 = layers.SeparableConv2D(128, (3, 3), activation='relu', padding='same', name='block3_conv1')(block2_pool)
block3_conv2 = layers.SeparableConv2D(128, (3, 3), activation='relu', padding='same', name='block3_conv2')(block3_conv1)
block3_conv3 = layers.SeparableConv2D(128, (3, 3), activation='relu', padding='same', name='block3_conv3')(block3_conv2)
block3_pool = layers.MaxPooling2D((2, 2), strides=(2, 2), name='block3_pool')(block3_conv3)

conv_output = layers.GlobalAveragePooling2D()(block3_pool)

dense1 = layers.Dense(200, activation='relu')(conv_output)
dense2 = layers.Dense(100, activation='relu')(dense1)
model_output = layers.Dense(5, activation='softmax')(dense2)
model = models.Model(img_input, model_output)
```
