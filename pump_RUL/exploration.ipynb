```python
import os, sys, glob
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('sensor.csv')
df = df.drop('Unnamed: 0', axis=1)
df.machine_status = (df.machine_status=='NORMAL').astype('int')
```

```python
df.timestamp = pd.to_datetime(df.timestamp)
```

```python
ax = df.plot(x='timestamp', y='sensor_01')
df.plot(x='timestamp', y='machine_status', ax=ax)
```
